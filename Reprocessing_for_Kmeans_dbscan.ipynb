```python
import pandas as pd
```

```python
df = pd.read_csv('data/reprocessed/edges.csv')
```

```python
ls_dict = []
for i, r in df.iterrows():
    d = dict()
    d['director'] = r['director']
    d[r['rating']] = 1
    ls_dict.append(d)
```

```python
ls = []
for i in ls_dict:
    mutual = [ k for k in ls_dict if i['director'] == k['director']]
    d = dict()
    for m in mutual:
        d.update(m)
    ls.append(d)

pd.DataFrame(ls).fillna(0).to_csv('data/reprocessed/director_features.csv', index=False)
```
